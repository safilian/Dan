# First approach for the problem - Prithivi


This code imports the SentenceTransformer and util modules from sentence_transformers. It defines a function library_sample_check_function_new_sentence_model that calculates the similarity between a title and a list of paragraphs. It uses the SentenceTransformer model 'all-mpnet-base-v2' to encode the title and paragraphs into vectors, then computes the cosine similarity between the title vector and each paragraph vector. The function returns a dictionary mapping each paragraph to its similarity score with the title.

```python
from sentence_transformers import SentenceTransformer, util

# Function to check the similarity of the title with the paragraphs
def library_sample_check_function_new_sentence_model(library_sample_check_list, title):
    model = SentenceTransformer('all-mpnet-base-v2')
    title_encoded = model.encode(title)
    paragraph_dict = {}
    for paragraph in library_sample_check_list:
        paragraph_encoded = model.encode(paragraph)
        similarity_score = round(util.pytorch_cos_sim(title_encoded, paragraph_encoded).item() * 100)
        paragraph_dict[paragraph] = similarity_score
    return paragraph_dict
```

```python
import time
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Function to get embeddings and calculate relevance scores
def get_relevance_scores(paragraphs, title, model="models/embedding-001"):
    result = genai.embed_content(
        model=model,
        content=paragraphs + [title],
        task_type="semantic_similarity"
    )

    # Extract vectors
    paragraph_vectors = np.array(result['embedding'][:-1])
    title_vector = np.array(result['embedding'][-1]).reshape(1, -1)

    # Calculate cosine similarities
    similarities = cosine_similarity(title_vector, paragraph_vectors)[0]

    # Convert to relevance scores (percentage)
    relevance_scores = similarities * 100

    return relevance_scores

# Read paragraphs from the text file
with open('paragraph_from_pdf.txt', 'r', encoding='utf-8') as file:
    paragraphs = [line.strip() for line in file]

```

```python
# Chunk size and rate limit
chunk_size = 5
requests_per_minute_limit = 60
delay_between_chunks = 60 / requests_per_minute_limit

# Process paragraphs in chunks
max_score = -1
min_score = 101
max_score_paragraph = ""
min_score_paragraph = ""

for i in range(0, len(paragraphs), chunk_size):
    # Get a chunk of paragraphs
    chunk = paragraphs[i:i + chunk_size]

    # Get relevance scores for the chunk
    relevance_scores = get_relevance_scores(chunk, title)

    # Update max and min scores and corresponding paragraphs
    for j, score in enumerate(relevance_scores):
        current_paragraph = chunk[j]
        if score > max_score:
            max_score = score
            max_score_paragraph = current_paragraph
        if score < min_score:
            min_score = score
            min_score_paragraph = current_paragraph

    # Pause to adhere to rate limit
    time.sleep(delay_between_chunks)

# Print results
print("Maximum Score:", max_score)
print("Paragraph with Maximum Score:", max_score_paragraph)
print("\nMinimum Score:", min_score)
print("Paragraph with Minimum Score:", min_score_paragraph)
```
